## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
#停止条件
i=0
while i<10
    print('hahha')
    i +=1
    #返回十次 i不写就会永远循环下去+=自加
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

```python
for i in range(0,10,2):
    print(i)
    #前闭后开的区间 用来计数
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range (10):
    if i==5:
        pass
    else:
        print(i)
```

```python
#最多三层
for i in range(10):
    for j in range(10):
        if i==5:
            contine
        print(i,j)
```

```python
for i in range(1,10):
    for j in range(1,i+1):
            print(i,'x',j,'=',i*j,end=' ')
    print()
    #执行完毕换行  所以print（）
```

```python

for num in range(10,20):  # 迭代 10 到 20 之间的数字
    for i in range(2,num): # 根据因子迭代
        if num%i == 0:      # 确定第一个因子
            j=num/i          # 计算第二个因子
            print '%d 等于 %d * %d' % (num,i,j)
            break            # 跳出当前循环
    else:                  # 循环的 else 部分
        print num, '是一个质数'
```
